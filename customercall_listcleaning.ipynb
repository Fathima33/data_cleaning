```python
import pandas as pd
```

Drop duplicate rows

```python
df.duplicated().value_counts()
```

```python
df=df.drop_duplicates()
```

```python
df.columns
```

```python
df.drop(columns="Not_Useful_Column",inplace=True)
```

```python
type(df.Last_Name[8])
```

```python
df.Last_Name=df.Last_Name.apply(lambda x:str(x))
```

```python
df.Last_Name=df.Last_Name.str.strip()
```

```python
df
```

```python
df.Last_Name=df.Last_Name.str.lstrip("\.../")
```

```python
df.Last_Name
```

```python
df['Last_Name']=df['Last_Name'].str.rstrip("_")
```

```python
df["Last_Name"]
```

```python
df['Phone_Number']=df["Phone_Number"].apply(lambda x:str(x))
```

```python
df['Phone_Number']=df['Phone_Number'].str.replace(r"\D","")
```

```python
df['Phone_Number']=df["Phone_Number"].apply(lambda x:x[0:3]+"-"+x[3:6]+"-"+x[6:])
```

```python

df['Phone_Number']
```

```python
df[["street","address","zipcode"]]=df.Address.str.split(",",2,expand=True)
```

```python
df
```

```python
df.columns
```

```python
df['Paying Customer']=df['Paying Customer'].replace("Yes","Y")
```

```python
df['Paying Customer']=df['Paying Customer'].replace("Y","Yes")
```

```python
df['Paying Customer']=df['Paying Customer'].replace("No","N")
df['Paying Customer']=df['Paying Customer'].replace("N","No")
```

```python
df
```

```python
df['Paying Customer']=df['Paying Customer'].replace("N/a","None")
```

```python
df.Do_Not_Contact=df.Do_Not_Contact.replace("No","N")
df.Do_Not_Contact=df.Do_Not_Contact.replace("N","No")
```

```python
df.Do_Not_Contact=df.Do_Not_Contact.replace("Yes","Y")
df.Do_Not_Contact=df.Do_Not_Contact.replace("Y","Yes")

```

```python
df=df[df.Do_Not_Contact=="No"]
```

```python
df
```

```python
df.reset_index()
```

```python
df['Phone_Number'] = df['Phone_Number'].replace("--", pd.NA)
```

```python
df.dropna(subset="Phone_Number",inplace=True)
```
