We are creating a function to assign nan value to the contract that consists On loan or Free values

```python

def ext_con_info(contract):
    if contract == 'Free' or 'On Loan' in contract:
        str_dt = np.nan
        en_dt = np.nan
        con_len = 0
    else:
        try:
            str_dt, en_dt = contract.split('~')
            str_yr = int(str_dt.strip())  # Added strip() to remove any leading or trailing spaces
            en_yr = int(en_dt.strip())
            con_len = en_yr - str_yr
        except:
            str_dt = np.nan
            en_dt = np.nan
            con_len = 0

    return str_dt, en_dt, con_len
```

```python
new_cols = ['Contract', 'Contract End', 'Contract Length(years)']

# Apply the function and split the results into separate columns
new_data = fifa['Contract'].apply(lambda x: pd.Series(ext_con_info(x)))

# Insert the new columns to the 'fifa' dataframe
for i in range(len(new_cols)):
    fifa[new_cols[i]] = new_data.iloc[:, i]
```

Convert the feet into inches

```python
df["Height"]=df["Height"].apply(lambda x:convert_height(x))
```

```python
def convert_height(x):
  if "cm" in x:
    return int(x.strip("cm"))
  else:
    feet,inches=x.split("'")
    cm=int(feet)*(30.48)+int(inches.strip('"'))*(2.54)
    return cm
```
